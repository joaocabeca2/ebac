```python
import pandas as pd

data = pd.read_json('deliveries.json',encoding='utf8')
data.head()
```

```python
data.info()
```

***dividir a latitude e longitude em duas colunas diferentes***

```python
lat_long = pd.json_normalize(data['origin'])
lat_long.head()
```

***combinar essas duas novas colunas de latitude e longitude ao dataframe original***

```python
dados_normalizados = pd.merge(left=data,right=lat_long,how='inner',left_index=True,right_index=True)
dados_normalizados.rename(columns={'lat':'latitude','lng':'longitude'},inplace=True)
dados_normalizados.head()
```

***excluir a coluna 'origin' que continha os dados de latitude e longitude nao normalizados***

```python
dados_normalizados = dados_normalizados.drop('origin',axis=1)
dados_normalizados.head()
```
